#Function for generating 3 new columns representing the three incorrect answers.
The function makes sure that the incorrect answers and the correct answer are all different

```python
# Function to generate unique incorrect answers
def generate_incorrect_answers(correct_answer, answer_column):
    incorrect_answers = []
    while len(incorrect_answers) < 3:
        incorrect_answer = correct_answer
        correct_length = len(correct_answer)

        if correct_length <= 3:
            while len(incorrect_answer) > 3 or incorrect_answer == correct_answer or incorrect_answer in incorrect_answers:
                incorrect_answer = np.random.choice(answer_column)
        elif 4 <= correct_length <= 8:
            while not (4 <= len(incorrect_answer) <= 8) or incorrect_answer == correct_answer or incorrect_answer in incorrect_answers:
                incorrect_answer = np.random.choice(answer_column)
        else:
            while len(incorrect_answer) <= 8 or incorrect_answer == correct_answer or incorrect_answer in incorrect_answers:
                incorrect_answer = np.random.choice(answer_column)
        incorrect_answers.append(incorrect_answer)

    return incorrect_answers


# Generate incorrect answers for each row
df['incorrect_answers'] = df.apply(lambda row: generate_incorrect_answers(row['Answer'], df['Answer']), axis=1)
df[['incorrect_answer_1', 'incorrect_answer_2', 'incorrect_answer_3']] = pd.DataFrame(df['incorrect_answers'].tolist(), index=df.index)
df.drop(columns=['incorrect_answers'], inplace=True)

df
```
